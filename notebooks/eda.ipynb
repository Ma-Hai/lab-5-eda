```python
#task A
import pandas as pd
df = pd.read_csv("../data/students (1).csv")
df.head()
```

```python
#task B

df.info()
missing_vals = df.isna().sum()
print('hi')
print(missing_vals)


```

GPA has the most missing values with a total of 5. This may be because some students haven't recieved grades yet.

```python
#task C
data = ['age','gpa']
summary = {
    'count': df[data].count(),
    'mean': df[data].mean(),
    'median': df[data].median(),
    'variance': df[data].var(),
    'SD': df[data].std(),
    'min': df[data].min(),
    'max': df[data].max()
}
summary = pd.DataFrame(summary)
print(summary)
```

```python
#task D
import matplotlib.pyplot as plt
import seaborn as sns

df['age'].hist(bins=range(14,19),align='left',rwidth=0.8)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title("Distrubution of Student Ages")
plt.show()

sns.histplot(df["gpa"],kde=True)
plt.xlabel('GPA')
plt.title("Distrubution of Student GPAs")
```

The plot made without SNS has lines providing a clearer view of what the y-value is.The plot made using SNS looks more aesthetic with a curve line showing the general flow of the data. 

```python
#task E

plt.scatter(df['age'],df['gpa'])
plt.xlabel('Age')
plt.ylabel('GPA')
plt.title('Age vs GPA')
plt.show()

correlation = df["age"].corr(df["gpa"])
print(correlation)
```

The correlation is closer to 1 so there is a relatively strong trend that as the age increases, so does the GPA.
